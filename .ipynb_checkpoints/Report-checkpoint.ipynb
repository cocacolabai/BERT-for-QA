## 1. Tokenization
輸入文本是一個個的句子，首先會在句子前面加上’[CLS]’，後面加上’[SEP]’。作業是以batch爲單位來tokenize的，因此還需要用到一個輸入input_masks，在中文的文本中，會拆成一個一個的字元，文本裡面包含英文的地方，會像一般英文tokenize一樣以單字為單位擷取（空格來分隔）



## 2. Answer span processing
原本我將每個char的轉換寫成一個chat_to _word_offset的vector來處理，但是後來我發現就算可以處理大部分的answer position，但是在英文單字變成[UNK]的case，會無法判斷原本tokenize之前的英文單字長度，會造成vector轉換的困難，所以後來使用bert裡面的encode函數處理＋在tokenize的list中搜尋子list，就可以成功找到tokenize後的answer span

## 3. Padding and Truncating
中文的max input token length是512，所以在context+question length超過512的case，將context後面拿掉，並把擷取後的context+question兩個list放到list中傳入model

## 5. Answer Length Distribution

````python
import json
import pandas as pd
import pylab as pl

with open('predict.json', 'r') as f:
    data = json.load(f)
answer_length=[]
for k in data.keys():
    answer_length.append(len(data[k]))

answer_length = pd.DataFrame(answer_length)
answer_length.hist(bins=31,cumulative=True,density=1)
pl.title("Cumulative Answer Length")
pl.xlabel("Length")
pl.ylabel("Count(%)")
````
